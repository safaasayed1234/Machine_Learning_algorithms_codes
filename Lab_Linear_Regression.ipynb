```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Dataset=pd.read_csv("/content/IQ.csv")
Dataset
```

```python
X=Dataset["Birthweight(X)"].values
Y=Dataset["IQ (Y)"].values
plt.scatter(X,Y)
plt.grid()
plt.title("child birthwieght vs. IQ")
plt.xlabel("birthwiegh")
plt.ylabel("IQ")
```

```python
x=X.reshape(-1, 1)
y=Y.reshape(-1, 1)
```

```python
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
# Create linear regression object
model = linear_model.LinearRegression()

# Train the model using the training sets
model.fit(x,y)
```

```python
w=model.coef_[0][0]
b=model.intercept_[0]
```

```python
y_pred=w*x+b
#y_pred=model.predict(x)
```

```python
plt.scatter(X,Y)
plt.plot(x,y_pred)
plt.grid()
plt.title("child birthwieght vs. IQ")
plt.xlabel("birthwiegh")
plt.ylabel("IQ")
```

```python
mean_squared_error(y_pred ,y)
```
